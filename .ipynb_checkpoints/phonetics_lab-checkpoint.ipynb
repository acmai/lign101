- read in and play sound files
- identify natural class features in the spectrogram

```python
import matplotlib.pyplot as plt
import numpy as np
import math
import parselmouth
import soundfile as sf
import IPython.display as ipd
```

```python
def draw_spectrogram(sound, figure_size=(15,7), dynamic_range=70):
    spectrogram = sound.to_spectrogram(maximum_frequency=8000)
    plt.figure(figsize=figure_size)
    X, Y = spectrogram.x_grid(), spectrogram.y_grid()
    sg_db = 10 * np.log10(spectrogram.values)
    plt.pcolormesh(X, Y, sg_db, vmin=sg_db.max() - dynamic_range, cmap='gray_r')
    plt.ylim([spectrogram.ymin, spectrogram.ymax])
    plt.xlabel("time [s]")
    plt.ylabel("frequency [Hz]")
    plt.show()
```

```python
snd = parselmouth.Sound('Voiceless_bilabial_plosive.wav')
draw_spectrogram(snd, dynamic_range=120)
ipd.Audio('Voiceless_bilabial_plosive.wav')
```
